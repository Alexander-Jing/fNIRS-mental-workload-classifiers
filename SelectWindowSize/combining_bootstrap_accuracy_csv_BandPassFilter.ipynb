```python
import numpy as np
import pandas as pd
import os
```

https://gist.github.com/michaelchughes/a9d4e55c5a043f75abdc24394c2ca58d#gistcomment-3857575

```python
def timestep_to_second_mapping(timesteps):
    if timesteps == 10.0:
        sec = 2.0
    elif timesteps == 25.0:
        sec = 5.0
    elif timesteps == 50.0:
        sec = 10.0
    elif timesteps == 100.0:
        sec = 20.0
    elif timesteps == 150.0:
        sec = 30.0
    elif timesteps == 200.0:
        sec = 40.0
    else:
        raise NameError('problem')
    
    return sec
```

### LR

```python
csv_folder_dir = './bootstrap_accuracy_csv_BandPassFilter/LR/'
```

```python
size2sec_df = pd.read_csv(os.path.join(csv_folder_dir, 'size2sec_bootstrap_accuracy.csv'))
size5sec_df = pd.read_csv(os.path.join(csv_folder_dir, 'size5sec_bootstrap_accuracy.csv'))
size10sec_df = pd.read_csv(os.path.join(csv_folder_dir, 'size10sec_bootstrap_accuracy.csv'))
size20sec_df = pd.read_csv(os.path.join(csv_folder_dir, 'size20sec_bootstrap_accuracy.csv'))
size30sec_df = pd.read_csv(os.path.join(csv_folder_dir, 'size30sec_bootstrap_accuracy.csv'))
size40sec_df = pd.read_csv(os.path.join(csv_folder_dir, 'size40sec_bootstrap_accuracy.csv'))
```

```python
accuracy_by_window_size_df = pd.concat([size2sec_df, size5sec_df, size10sec_df, size20sec_df, size30sec_df, size40sec_df], ignore_index=True)
```

```python
window_size_seconds = [timestep_to_second_mapping(i) for i in accuracy_by_window_size_df.window_size.values]
```

```python
window_size_seconds
```

```python
accuracy_by_window_size_df['window_size_sec'] = window_size_seconds
```

```python
accuracy_by_window_size_df
```

```python
accuracy_by_window_size_df.to_csv('./bootstrap_accuracy_csv_BandPassFilter/LR/accuracy_by_window_size.csv', index=False)
```

```python
pd.read_csv('./bootstrap_accuracy_csv_BandPassFilter/LR/accuracy_by_window_size.csv')
```
